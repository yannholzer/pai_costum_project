```python
df = data.get_dataframe_single_file(data_path)
df_disk, df_planet = data.get_disk_planets_dataframe(df)
```

```python
try:
    plt.style.use("~/Documents/uni/phd/mypackage/mypackage/pyplot/presentation.mlpstyle")
except:
    pass

# plot the disk features distribution


columns = df_disk.columns[1:] # remove system number


bins=50
color="dodgerblue"

fig, axes = plt.subplots(1, 5, figsize=(20, 5))
for i_d, d_f in enumerate(columns):
    ax = axes[i_d]
    ax.hist(df_disk[d_f], bins=bins, color=color)
    ax.set(xlabel=d_f)
    if i_d == 0:
        ax.set(ylabel="count")
        

plt.tight_layout()
plt.savefig("rapport/images/disk_features_distribution.png")
plt.show()

fig, axes = plt.subplots(1, 5, figsize=(20, 5))
for i_d, d_f in enumerate(columns):
    ax = axes[i_d]
    ax.hist(np.log10(df_disk[d_f]), bins=bins, color=color)
    ax.set(xlabel=d_f)
    if i_d == 0:
        ax.set(ylabel="count")
        

plt.tight_layout()
plt.savefig("rapport/images/disk_features_log_distribution.png")
plt.show()
```
